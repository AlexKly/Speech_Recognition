```python
import numpy as np
from scipy.signal import medfilt
import wavio

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_roc_curve
from sklearn.preprocessing import normalize

from python_speech_features import mfcc, delta

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mfcc_features = np.loadtxt('mfcc_features.txt')

labels = np.zeros((mfcc_features.shape[0]))
data_feat = np.zeros((mfcc_features.shape[0], mfcc_features.shape[1] - 1))
for i in range(mfcc_features.shape[0]):
    labels[i] = mfcc_features[i][0]
    data_feat[i][:] = mfcc_features[i][1:] + np.random.normal(0, 10, mfcc_features.shape[1] - 1)
    
#data_feat = normalize(data_feat)

X_train, X_test, y_train, y_test = train_test_split(data_feat, labels, train_size=0.75, random_state=42)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(7), max_iter=10000, activation='relu', learning_rate='invscaling')
#clf = MLPClassifier(hidden_layer_sizes=(15, 15, 15, 15), max_iter=10000, activation='relu')
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
plot_roc_curve(clf, X_test, y_test)
```

```python
# 90:
path = 'C:/Users/Alex/Desktop/VAD/'
filename = 'Test_1_2.wav'
wav = wavio.read(path + filename)
sound = wav.data
sample_rate = wav.rate

mfcc_feat = mfcc(sound, sample_rate, nfft=512, appendEnergy=False)
#mfcc_feat -= (np.mean(mfcc_feat, axis=0) + 1e-8)
#mfcc_feat = normalize(mfcc_feat)

labels = clf.predict(mfcc_feat)
```

```python
start_frame = 0
end_frame = int(sample_rate * 0.025)
step_frame = int(sample_rate * 0.01)
num_frames = mfcc_feat.shape[0]

X_coor = np.zeros(num_frames)
Y_coor = np.zeros(num_frames)
Marker = np.zeros(num_frames)
X_coor[0] = start_frame
Y_coor[0] = end_frame

for i in range(1, num_frames):
    
    start_frame += step_frame
    end_frame += step_frame
    X_coor[i] = start_frame
    Y_coor[i] = end_frame
    Marker[i] = labels[i]
    
Marker = medfilt(labels, 21)
```

```python
plt.figure(num=None, figsize=(16, 9), dpi=80, facecolor='w', edgecolor='k')
for i in range(num_frames):
    if labels[i] == 1:
        color_bar = 'red'
    else:
        color_bar = 'white'
    plt.plot([X_coor[i], X_coor[i]], [-1, 1], color=color_bar)
    plt.plot([X_coor[i], Y_coor[i]], [1, 1], color=color_bar)
    plt.plot([Y_coor[i], Y_coor[i]], [-1, 1], color=color_bar)
    
plt.plot(sound / max(abs(sound)), color='blue')
```

```python
list_ind = []
list_ind_loc = []
list_cnt = []
cnt = 0
for i in range(len(labels) - 1):
    if labels[i] == 1.0:
        cnt += 1
        list_ind_loc.append(i)
        if labels[i + 1] == 0.0:
            list_cnt.append(cnt)
            list_ind.append(list_ind_loc)
            list_ind_loc = []
            cnt = 0

marks = labels
for i in range(len(list_cnt)):
    if list_cnt[i] < 5:
        indexes = list_ind[i]
        for index in indexes:
            marks[index] = 0.0
            
list_ind_zero = []
list_ind_zero_loc = []
list_cnt_zero = []
cnt_zero = 0
for i in range(len(marks) - 1):
    if marks[i] == 0.0:
        cnt_zero += 1
        list_ind_zero_loc.append(i)
        if marks[i + 1] == 1.0:
            list_cnt_zero.append(cnt_zero)
            list_ind_zero.append(list_ind_zero_loc)
            list_ind_zero_loc = []
            cnt_zero = 0
            
for i in range(len(list_cnt_zero)):
    if list_cnt_zero[i] < 10:
        indexes_zero = list_ind_zero[i]
        for index_zero in indexes_zero:
            marks[index_zero] = 1.0
```
