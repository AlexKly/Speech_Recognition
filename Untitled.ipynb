```python
import numpy as np
import scipy.special

from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from pandas import DataFrame
```

```python
class MultiLayers_Classifier:
    
    def __init__(self, num_neurons_in_layer, learning_rate, epochs):
        # Number of neurons in layers:
        self.num_neurons_in_layer = num_neurons_in_layer
        # Learning rate:
        self.learning_rate = learning_rate
        # Number of learing loops:
        self.epochs = epochs
        
        # Initialization activation fucntion:
        self.activation_function = lambda x: scipy.special.expit(x)    
        # Number layers:
        self.num_layers = len(num_neurons_in_layer)
        # Initialization weights:
        self.Weigths = []
        
        # Generation layers of weigths:
        for layer in range(self.num_layers - 1):
            next_layer = num_neurons_in_layer[layer + 1]
            pres_layer = num_neurons_in_layer[layer]
            weights_layer = np.random.randn(next_layer, pres_layer) * np.sqrt(2 / (pres_layer + next_layer))
            self.Weigths.append(weights_layer)
                
    # Fit classifier:
    def fit_classifier(self, inputLayer_train, label):
        train_list = inputLayer_train
        num_samples = train_list.shape[0]
        targets_list = label
        for epoch in range(self.epochs):
            for sample in range(num_samples):
                inputLayer_values = train_list[sample]
                target = np.zeros(train_list.shape[1]) + 0.01
                target[int(targets_list[sample])] = 0.99
                Output_values = [inputLayer_values]
                
                # Forward propagation:
                for layer in range(self.num_layers - 1):
                    outputs = Output_values[layer]
                    new_outputs = self.activation_function(np.dot(self.Weigths[layer], outputs))
                    Output_values.append(new_outputs)
                
                Errors_values = []
                # Get errors:
                for layer in range(self.num_layers - 1):
                    if layer == 0:
                        errors = target - Output_values[-1]
                    else:
                        errors = np.dot(self.Weigths[-layer].T, Errors_values[layer - 1])
                    Errors_values.append(errors)
                    
                # Back propagation:
                for layer in range(self.num_layers - 1):
                    self.Weigths[-layer - 1] += self.learning_rate * np.dot(Errors_values[layer] * Output_values[-layer - 1] * (1 - Output_values[-layer - 1]), np.transpose(Output_values[-layer - 2]))
                
                
        return Output_values, Errors_values
                
    
    #def fit_classifier(self, inputLayer_train):
    #    train_list = inputLayer_train
    #    #targets_list = label
    #    num_samples = train_list.shape[0]
    #    for sample in range(num_samples):
    #        #if targets_list[sample] == 0:
    #        #    targets_list = [1, 0]
    #        #else:
    #        #    targets_list = [0, 1]
    #        inputLayer_values = train_list[sample]
    #        for epoch in range(self.epochs):
    #            # Forward propagation:
    #            Output_values = [inputLayer_values]
    #            for layer in range(self.num_layers - 1):
    #                outputs = Output_values[layer]
    #                outputs = np.insert(outputs, 0, 1.0)
    #                new_outputs = self.activation_function(np.dot(self.Weigths[layer], outputs))
    #                Output_values.append(new_outputs)
    #            
    #            # Get errors:
    #            Layers_errors = []
    #            for layer in range(self.num_layers - 1):
    #                if layer == 0:
    #                    error = -Output_values[-1] * (1 - Output_values[-1]) * (targets_list - Output_values[-1])
    #                else:
    #                    outputs = Output_values[-layer - 1]
    #                    outputs = np.insert(outputs, 0, 1.0)
    #                    error = -outputs * (1 - outputs) * np.dot(self.Weigths[-layer].T, Layers_errors[layer - 1])
    #                    error = error[1:]
    #                Layers_errors.append(error)
    #        
    #            # Back propagation:
    #            for layer in range(self.num_layers - 1):
    #                errors = Layers_errors[layer]
    #                #errors = np.insert(errors, 0, 0.0)
    #                outputs = Output_values[-layer - 1]
    #                #outputs = np.insert(outputs, 0, 1.0)
    #                self.Weigths[-layer - 1] += -self.learning_rate * np.dot(errors, outputs)
            
        #return Output_values#, Layers_errors
            
    #def predict(self, inputLayer_test):
    #    answers = []
    #    data_test = inputLayer_test
    #    num_samples = data_test.shape[0]
    #    for sample in range(num_samples):
    #        inputLayer_test = data_test[sample]
    #        Input_layer = [inputLayer_test]
    #        Output_layer = [inputLayer_test]
    #        for layer in range(self.num_layers - 1):
    #            outputs = Output_layer[layer]
    #            outputs = np.insert(outputs, 0, 0.1)
    #            new_inputs = np.dot(self.Weigths[layer], outputs)
    #            new_outputs = self.activation_function(new_inputs)
    #            Output_layer.append(new_outputs)
    #        
    #        answers.append(Output_layer[-1])
    #        
    #    return answers
                     
```

```python
X, y = make_blobs(n_samples=10000, centers=2, n_features=2)

df = DataFrame(dict(x=X[:, 0], y=X[:, 1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show
```

```python
X = normalize(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
num_samples, num_features = X_train.shape
num_classes = num_features
num_hidden_neurons = 5
lr = 0.01
epochs = 100

clf = MultiLayers_Classifier([num_features, num_hidden_neurons, num_classes], lr, epochs)
```

```python
clf.fit_classifier(X_train, y_train)
```

```python
labels = clf.predict(X_test)
```

```python
list_labels = []
for i in range(len(labels)):
    if labels[i][0] > labels[i][1]:
        list_labels.append(0)
    else:
        list_labels.append(1)
```
